```python
demographic_df = pd.read_csv('../data_files/metro_demographic_by_city_state.csv', index_col=[0], header=0)
BLS = pd.read_csv('../data_files/bls_metro_2020_clean.csv', index_col=[0], header=0)
```

```python
search_area = "Beckley, WV"
```

```python
all_occupation_row_df = BLS[(BLS['occ_title']=='All Occupations') & (BLS['area_title']==search_area)]
all_occupation_row_df
```

```python
area_df = BLS[(BLS['area_title']==search_area) & (BLS['occ_title']!='All Occupations') ]
area_df.sort_values(by='tot_emp', ascending=False).head(20)
```

```python
area_major_df = BLS[(BLS['area_title']==search_area) & (BLS['occ_title']!='All Occupations') & (BLS['o_group']=='major') ]
area_major_df.sort_values(by='tot_emp', ascending=False).head()
```
