## Prepare BLS_df with only these BLS columns

```python
BLS_df = BLS[['area_title', 'occ_title', 'tot_emp', 'jobs_1000']]
BLS_df['tot_emp'] = BLS_df['tot_emp'].astype(int)

# Drop the rows where occ_title is 'All Occupations'
BLS_df = BLS_df[BLS_df['occ_title']!='All Occupations']
BLS_df.head(100)
```

## Define occ_title group

These are our predefined group that we can pass into PostGres later. For now we pass into our **isin** filter next step.

For final presentation, these job_groups can be a select drop-down to see the same analysis run against several different job categories.

I scrolled through all the jobs list and manually read and extracted the occ_titles that fits "technical". I will revisit this.

```python
job_groups = {
    'technical' : ['Management Occupations','Computer and Information Systems Managers', 'Computer and Mathematical Occupations', 'Computer Systems Analysts','Computer User Support Specialists', 'Network and Computer Systems Administrators', 'Software Developers and Software Quality Assurance Analysts and Testers', 'Computer Occupations, All Other', 'Computer Hardware Engineers', 'Information Security Analysts', 'Computer Network Support Specialists', 'Computer Network Architects', 'Database Administrators and Architects', 'Computer Programmers', 'Web Developers and Digital Interface Designers', 'Data Scientists and Mathematical Science Occupations, All Other', 'Computer Numerically Controlled Tool Operators', 'Computer Numerically Controlled Tool Programmers', 'Computer and Information Research Scientists' ],
    'healthcare' : ['Registered Nurses', 'Nurse Practitioners', 'just examples']
}

job_groups['technical']
```

## Flag BLS row as istech if occ_title isin list

Creates a filtered DF where occ_title is in our Titles list defined above

```python
#filteredBLS_df = BLS_df.loc[BLS_df['occ_title'].isin(job_groups['technical'])]
filteredBLS_df = BLS_df
filteredBLS_df['istech'] = np.where(filteredBLS_df["occ_title"].isin(job_groups['technical']), 1, 0)
filteredBLS_df
```

### Group by area_title

#### Note: Only including jobs_1000 here to keep column ordering, will calculate actual jobs_1000 values below

```python
merged_grouped = filteredBLS_df.groupby(['area_title', 'istech'], as_index=False).agg(
    area_title = ('area_title', lambda x: x.iloc[0]),
    tot_emp = ('tot_emp', 'sum'),
    jobs_1000 = ('jobs_1000', lambda x: 0)
    )
```

# Merge the filtered BLS with the Demographic DF on 'area_title'

This creates a DataFrame matching all merged_grouped['area_title'] to demographic_df on 'area_title'


```python
tech_not_tech_df = pd.merge(merged_grouped, demographic_df, on=['area_title'])
```

## Calculate jobs_1000

```python
tech_not_tech_df['jobs_1000'] = round(( tech_not_tech_df['tot_emp'] / tech_not_tech_df['population']) * 1000).astype(int)

# Drop this column, using area_title
tech_not_tech_df = tech_not_tech_df.drop(columns=['area_name'])
```

## Analyzing employment data as a whole

**Just Brainstorming while going through this data**

How we can use this dataset is lets say, we have "family_poverty" as the target, and the demographic and jobs_1000 and tot_emp as the independent variables. If we find a pattern with high accuracy to predict "family_poverty", then we can see which variable will help reduce "family_poverty". Could be reduce the median_age, or something as small as increasing the "average_income" by 2k, can reduce "family_poverty" by n%. Or increase the number of High School graduates.

```python
# setting jobs_1000 as 0 to keep column order for human viewing
merged_grouped_all_df = tech_not_tech_df.groupby(['area_title'], as_index=False).agg(
    tot_emp = ('tot_emp', 'sum'),
    jobs_1000 = ('jobs_1000', lambda x: 0)
)

merged_grouped_all_df = pd.merge(merged_grouped_all_df, demographic_df, on=['area_title'])
merged_grouped_all_df['jobs_1000'] = round(( merged_grouped_all_df['tot_emp'] / merged_grouped_all_df['population']) * 1000).astype(int)
merged_grouped_all_df.head(10)
```
